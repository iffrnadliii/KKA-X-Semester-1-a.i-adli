menghitung luas persegi panjang

```python
panjang = int(input("masukkan panjang dari persegi panjang (cm): "))
lebar = int(input(" masukkan lebar dari persegi panjang (cm): "))

luas = panjang * lebar

print(f" \nluas persegi panjang yaitu {luas} cm^2 ")
```

menghitung volume bola

```python
jari_jari = int(input("masukkan jari_jari dari bola (cm): "))

volume_bola = 4/3 * 22/7 * jari_jari

print(f"\nvolume_bola yaitu {volume_bola} cm^3 ")
```
